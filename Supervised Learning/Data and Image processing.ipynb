```python
import os
import re
import pandas as pd
import rasterio
from rasterio.windows import Window
import numpy as np
from PIL import Image
from datetime import datetime
from tqdm.auto import tqdm
from sklearn.cluster import DBSCAN

TIFF_NAME = "Quebec_MAG_DV1_TIFF.tif"
SAMPLES_NAME = "geology_ml_training_data.csv"

def find_path(name, is_dir=False):
    print(f"Searching for {'directory' if is_dir else 'file'}: {name}...")
    for root, dirs, files in os.walk('/content/drive/MyDrive/SIADS696 - Milestone II'):
        if is_dir:
            if name in dirs and 'processed' in root:
                path = os.path.join(root, name)
                print(f"Found directory: {path}")
                return path
        else:
            if name in files:
                path = os.path.join(root, name)
                print(f"Found file: {path}")
                return path
    print(f"Not found")
    return None

def calculate_window_pixels(tiff, km_size=5):
    res_x, res_y = tiff.res
    return int((km_size * 1000) / res_x), int((km_size * 1000) / res_y)

def extract_and_save_patch(tiff, easting, northing, idx, pix_x, pix_y, output_dir, final_size=(600, 600)):
    try:
        row, col = tiff.index(easting, northing)
        win = Window(col - pix_x//2, row - pix_y//2, pix_x, pix_y)
        image_data = tiff.read(window=win)
        if image_data.size == 0: return False

        if tiff.count >= 3:
            image_data = np.transpose(image_data[:3, :, :], (1, 2, 0))
        else:
            image_data = image_data[0, :, :]

        # Discard images with white ratio bigger than 0.8
        if np.mean(image_data >= 250) > 0.8:
          return False

        # img_min, img_max = np.nanmin(image_data), np.nanmax(image_data)
        # if img_max > img_min:
        #     image_data = ((image_data - img_min) * (255.0 / (img_max - img_min))).astype(np.uint8)
        # else:
        #     image_data = np.zeros(image_data.shape, dtype=np.uint8)

        img = Image.fromarray(image_data)
        # if img.size != final_size:
        #     img = img.resize(final_size, resample=Image.LANCZOS)

        img.save(os.path.join(output_dir, f"{idx}_native_res.jpg"), quality=90)
        return True
    except Exception: return False

def generate_images(df, row_range=None, km_size=5, output_dir='', output_pixels=600):
    tiff_path = find_path(TIFF_NAME)
    x_col, y_col, id_col = 'Easting', 'Northing', 'Sample_ID'
    if row_range is None:
        row_start, row_end = (0, len(df))
    else:
        row_start, row_end = row_range

    print(f"Starting generation: {row_end - row_start} images at {km_size}km scale...")
    with rasterio.open(tiff_path) as tiff:
        src_pix_x, src_pix_y = calculate_window_pixels(tiff, km_size=km_size)
        print(src_pix_x, src_pix_y)
        current = row_start
        for idx, row in tqdm(df[row_start:row_end].iterrows(), total=row_end - row_start, desc="Processing Images"):
            df.loc[idx, 'hasImage'] = extract_and_save_patch(tiff, row[x_col], row[y_col], row['spatial_cluster'], src_pix_x, src_pix_y, output_dir, (output_pixels, output_pixels))
            current += 1
    print(f"\nCompleted! Generated {current} images in: {output_dir}")
    return df
```

```python
def process_dataset(cluster_proximity=0):
    samples_path = find_path(SAMPLES_NAME)
    df = pd.read_csv(samples_path, low_memory=False)

    ## Filling missing NaNs with median and adding missing column
    mineral_columns = ['AU','AG','CU','CO','NI']
    df['STRAT'] = df['STRAT'].fillna('')
    for col in mineral_columns:
      df[f"{col}_missing"] = df[col].isna().astype("int8")
      median_value = df[col].median()
      df[col] = df[col].fillna(median_value)

    ## Setting up clusters of rows with close by coordinates
    if cluster_proximity > 0:
      coords = df[['Easting', 'Northing']].values
      clustering = DBSCAN(eps=cluster_proximity, min_samples=1).fit(coords)
      df['spatial_cluster'] = clustering.labels_

    ## Grouping by rows of same clusters
    # Concatenating string columns and doing Mean for all other columns
    agg_dict = {'CODE_LITH': lambda x: '-'.join(x.astype(str)),
                'STRAT': lambda x: '-'.join(x.astype(str))}
    for col in df.columns:
        if col not in ['spatial_cluster', 'CODE_LITH', 'STRAT']:
            agg_dict[col] = 'mean'
    df = df.groupby('spatial_cluster', as_index=False).agg(agg_dict)

    ## One hot encoding CODE_LITH
    df['CODE_LITH'] = df['CODE_LITH'].str.replace(r'/', '-', regex=True)
    df['CODE_LITH'] = df['CODE_LITH'].str.replace(r'\(', '-', regex=True)
    df['CODE_LITH'] = df['CODE_LITH'].str.replace(r'\)(?=-|$)', '', regex=True)
    df['CODE_LITH'] = df['CODE_LITH'].str.replace(r'\)(?!-)', '-', regex=True)

    lith_ohe = df['CODE_LITH'].str.get_dummies(sep='-')
    df = pd.concat([df, lith_ohe], axis=1)

    # Removing one hot encoded columns with less than 5 ocurrences
    rare_cols = [col for col in lith_ohe.columns if df[col].sum() < 5]
    df = df.drop(columns=rare_cols)

    ## Sort by coordinates
    df = df.sort_values(by=['Easting', 'Northing']).reset_index(drop=True)
    df['hasImage'] = False

    ## Save processed tabular data
    output_processed_path = '/content/drive/MyDrive/SIADS696 - Milestone II/data/processed/tabular/processed('+str(cluster_proximity)+')_'+SAMPLES_NAME
    df.to_csv(output_processed_path, index=False)

    return df
```
